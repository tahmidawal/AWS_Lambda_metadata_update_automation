```python
import pandas as pd
import googlemaps
import time

API_KEY = 'AIzaSyAjLS_v_OIDJCYnD0-mPhZNB_BMSyDhSOo'
map_client = googlemaps.Client(API_KEY)

def get_place_info(location_name):
    all_results = []
    try:
        response = map_client.places(query=location_name)
        all_results.extend(response.get('results', []))

        while 'next_page_token' in response:
            time.sleep(2)
            response = map_client.places(query=location_name, page_token=response['next_page_token'])
            all_results.extend(response.get('results', []))

        return all_results
    except Exception as e:
        print(f"Error: {e}")
        return []

# Read the list of cities from the text file
with open('us_city_state_names.txt', 'r') as file:
    cities = file.read().splitlines()

df = pd.DataFrame(columns=['City', 'Location', 'Name', 'Address', 'Place ID', 'Latitude', 'Longitude', 'Rating', 'User Ratings Total', 'Website', 'Phone Number', 'Description'])

for city in cities:
    base_queries = ['hospitals', 'clinics', 'diagnostic centers']
    for query in base_queries:
        full_query = f"{query} in {city}"
        places = get_place_info(full_query)
        for place in places:
            place_details = map_client.place(place_id=place.get('place_id', ''))['result']
            location = place.get('geometry', {}).get('location', {})
            df = df.append({
                'City': city,
                'Location': full_query,
                'Name': place.get('name', ''),
                'Address': place.get('formatted_address', ''),
                'Place ID': place.get('place_id', ''),
                'Latitude': location.get('lat', ''),
                'Longitude': location.get('lng', ''),
                'Rating': place.get('rating', ''),
                'User Ratings Total': place.get('user_ratings_total', ''),
                'Website': place_details.get('website', ''),
                'Phone Number': place_details.get('formatted_phone_number', ''),
                'Description': ', '.join(place.get('types', []))
            }, ignore_index=True)

output_file_path = '/mnt/data/healthcare_orgs_usa.csv'
df.to_csv(output_file_path, index=False)

print("Data collection complete. File saved to:", output_file_path)
```

```python
import pandas as pd
import folium

# Load the CSV file
data = pd.read_csv('healthcare_orgs_usa.csv')

# Create a map object, starting at a central location
mymap = folium.Map(location=[37.0902, -95.7129], zoom_start=4)  # USA central coordinates

# Loop through the data and add markers
for idx, row in data.iterrows():
    popup_text = f"{row['Name']}<br>Rating: {row['Rating']}<br>{row['Address']}"
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_text
    ).add_to(mymap)

# Save or display the map
mymap.save('healthcare_orgs_map.html')  # Change the path as needed
```
